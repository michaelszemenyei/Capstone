```python
# Changing the order of the colums
away = pd.DataFrame(away, columns = ['A_FTPct', 'A_EFGPct', 'A_ThreePARt', 'A_FTR', 
           'A_REBPct', 'A_BLKPct', 'A_AST_TOV_Ratio', 'Target',])
```

```python
# Creating a function to bin the target variable
def target_binning(Target):
    if (Target) == 0:  # Will return 0 if a team loses the game
        return(1)
    else:                   # Will return 1 if a team wins the game
        return(0)
```

```python
# Applying the function in order to bin the target variable 
away['New_Target'] = away['Target'].apply(target_binning)
```

```python
away.head()
```

```python
#Dropping the Old Target Variable
away = away.drop(['Target'], axis=1)
```

```python
away.head()
```

```python
# Renaming the columns in order get ride of the % signs 
away.rename(columns={'New_Target':'Target'}, inplace=True)
```
