```python
# import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read the csv file 
df = pd.read_csv('../csv_files/Capstone_p1_iqr.csv', index_col=0)
df.head()
```

```python
# Getting a list of column names 
df.columns
```

```python
# import the preprocessing package from sklearn
from sklearn import preprocessing

# define column names
cols = ['H_FTPct', 'H_EFGPct', 'H_ThreePARt', 'H_FTR', 'H_REBPct', 'H_BLKPct', 
        'H_AST_TOV_Ratio', 'A_FTPct', 'A_EFGPct', 'A_ThreePARt', 'A_FTR', 
        'A_REBPct', 'A_BLKPct', 'A_AST_TOV_Ratio']

# scale all columns to have a range of 0-1
x = df.values
scaler = preprocessing.MinMaxScaler()
x_scaled = scaler.fit_transform(x)
df_scaled = pd.DataFrame(x_scaled,columns=cols)
df_scaled.head()
```

```python
# verify that the min and max values for each column are 0 and 1 respectively 
df_scaled.describe()
```

```python
# export new dataframe to csv
df_scaled.to_csv('../csv_files/Capstone_p1_minmax.csv')
```
